```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np

iris = load_iris()
iris.keys()
iris_data = iris.data
iris_target = iris.target
df_iris = pd.DataFrame(data = iris_data, columns=iris.feature_names)
```

```python
df_iris
```

```python
df_iris.columns
```

```python
df_target = pd.DataFrame(data = iris_target, columns=['label'])
data = pd.concat([df_iris, df_target], axis=1)
```

```python
data
```

```python
X = data.iloc[:,0:3].values
y = data.iloc[:,3].values


from sklearn.model_selection import train_test_split
X_train,X_test, y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0) 
```

```python
X
```

```python
y
```

```python
mean = X_train.mean(axis=0)
X_train -= mean
std = X_train.std(axis=0)
X_train /= std

X_test -= mean
X_test /= std
```

```python
from keras.models import Model
from keras import layers
from keras.layers import Input, Dense

def build_model(): 
    inputs = Input(shape=(3,))

    # 모델 구조 정의
    x = Dense(64, activation='relu',input_shape=(X_train.shape[1],))(inputs)
    x2 = Dense(64, activation='relu')(x)
    outputs = Dense(1)(x2)

    # 모델 정의
    model = Model(inputs=inputs, outputs=outputs)
    model.compile(optimizer='rmsprop', loss='mse', metrics=['mae'])
    return model

model = build_model()
model.summary()
```

```python
model.fit(X_train, y_train, epochs=80, batch_size=16, verbose=0)
test_mse_score, test_mae_score = model.evaluate(X_test, y_test)
```
